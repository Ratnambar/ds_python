```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        new_node = Node(data)
        if self.root == None:
            self.root = new_node
            return self.root
        else:
            st = []
            st.append(self.root)
            while len(st) > 0:
                curr = st.pop(0)
                if curr.left:
                    st.append(curr.left)
                else:
                    curr.left = new_node
                    return self.root
                if curr.right:
                    st.append(curr.right)
                else:
                    curr.right = new_node
                    return self.root

    def find_maximum(self):
        if self.root == None:
            return 
        else:
            st = []
            st.append(self.root)
            m = -10000
            while len(st) > 0:
                curr = st.pop(0)
                m = max(m, curr.data)
                if curr.left:
                    st.append(curr.left)
                if curr.right:
                    st.append(curr.right)
        return m

    def search_element(self, data):
        if self.root == None:
            return
        if self.root.data == data:
            return True
        else:
            st = []
            st.append(self.root)
            while len(st) > 0:
                curr = st.pop(0)
                if curr.data == data:
                    return True
                if curr.left:
                    st.append(curr.left)
                if curr.right:
                    st.append(curr.right)
            return False
    
    def size_of_tree(self):
        if self.root == None:
            return 0
        if self.root and (self.root.left == None and self.root.right == None):
            return 1
        else:
            st = []
            st.append(self.root)
            count = 0
            while len(st) > 0:
                curr = st.pop(0)
                count+=1
                if curr.left:
                    st.append(curr.left)
                if curr.right:
                    st.append(curr.right)
        return count

    def level_order_in_reverse_order(self):
        if self.root == None:
            return 0
        if self.root and (self.root.left == None and self.root.right == None):
            return 1
        else:
            st = []
            st.append(self.root)
            res_st = []
            while len(st) > 0:
                curr = st.pop(0)
                res_st.append(curr.data)
                if curr.left:
                    st.append(curr.left)
                if curr.right:
                    st.append(curr.right)
                
            while len(res_st) > 0:
                res = res_st.pop(0)
                print(res,end=" ")

    def height(self):
        if self.root == None:
            return 0
        if(self.root) and (self.root.left == None and self.root.right == None):
            return 1
        else:
            st = []
            count = 0
            st.append(self.root)
            while len(st) > 0:
                for _ in range(len(st)):
                    curr = st.pop(0)
                    if curr.left:
                        st.append(curr.left)
                    if curr.right:
                        st.append(curr.right)
                count+=1
            return count
    
    def deepest_node(self):
        if self.root == None:
            return None
        else:
            st = []
            st.append(self.root)
            while len(st) > 0:
                curr = st.pop(0)
                if curr.left:
                    st.append(curr.left)
                if curr.right:
                    st.append(curr.right)
            print(curr.data)

    def find_number_of_leafs(self):
        if self.root == None:
            return
        else:
            st = []
            st.append(self.root)
            count = 0
            while len(st) > 0:
                curr = st.pop(0)
                if curr.left == None and curr.right == None:
                    count+=1
                else:
                    if curr.left:
                        st.append(curr.left)
                    if curr.right:
                        st.append(curr.right)
            return count
                    
    def level_of_maximum_sum(self):
        if self.root == None:
            return [0,0]
        else:
            st = []
            st.append(self.root)
            level = 1
            max_level = 0
            max_sum = float('-inf')
            while len(st) > 0:
                p_sum = 0
                for i in range(len(st)):
                    curr = st.pop(0)
                    p_sum+=curr.data
                    if curr.left:
                        st.append(curr.left)
                    if curr.right:
                        st.append(curr.right)
                if max_sum < p_sum:
                    max_sum = p_sum
                    max_level = level
                level+=1
                print(max_level,max_sum)
            return [max_level, max_sum]
    def find_ancestors_of_a_node(self, data):
        if self.root == None:
            return None
        else:
            st = []
            st.append((self.root,[]))
            d = dict()
            while len(st) > 0:
                curr,path = st.pop(0)
                if curr.data == data:
                    return path
                if curr.left:
                    st.append((curr.left, path+[curr.data]))
                if curr.right:
                    st.append((curr.right, path+[curr.data]))
            return path

    
    def inorder_traversal(self, node):
        if node:
            self.inorder_traversal(node.left)
            print(node.data, end=' ')
            self.inorder_traversal(node.right)
                

bt = BinaryTree()
n = int(input("Enter the number of elements in the binary tree: "))
for _ in range(n):
    data = int(input("Enter data: "))
    bt.insert(data)

print("Inorder traversal of the binary tree:")
bt.inorder_traversal(bt.root)
print()
# print("largest element is : ",bt.find_maximum())
# print(bt.search_element(int(input("enter element for search"))))
# print("size of binary tree: ",bt.size_of_tree())
# bt.level_order_in_reverse_order()
# bt.height()
# bt.deepest_node()
# print("number of leaf node is : ",bt.find_number_of_leafs())
# print(bt.level_of_maximum_sum())
bt.find_ancestors_of_a_node(int(input("enter node for finding ancestors")))
```
