```python
# sort linked list using merge sort
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        if self.head == None:
            self.head = Node(data)
        else:
            new_node = Node(data)
            p = self.head
            while p.next != None:
                p = p.next
            p.next = new_node
            p = new_node
        return self.head
    
    def middle(self, head):
        slow = head
        fast = head.next
        while fast is not None and fast.next is not None:
            fast = fast.next.next
            slow = slow.next
        return slow
    
    def merge(self, left, right):
        if left == None:
            return right
        if right == None:
            return left
        else:
            res = None
            if left.data <= right.data:
                res = left
                res.next = self.merge(left.next, right)
            else:
                res = right
                res.next = self.merge(left, right.next)
            return res

    def merge_sort(self, head):
        if head is None or head.next is None:
            return head
        else:
            mid = self.middle(head)
            next_to_mid = mid.next
            mid.next = None

            left = self.merge_sort(head)
            right = self.merge_sort(next_to_mid)

            sorted_list = self.merge(left, right)
            return sorted_list
        
    def sort(self):
        self.head = self.merge_sort(self.head)

    def display(self):
        if self.head == None:
            return "empty list."
        else:
            p = self.head
            while p != None:
                print(p.data,end="->")
                p = p.next
            print("None")

ll = LinkedList()
for i in range(int(input("length of linked list"))):
    ll.insert_at_end(int(input("enter linked list element")))
ll.sort()
ll.display()
```

```python
# check if a linked list is palindrom or not
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        if self.head == None:
            self.head = Node(data)
        else:
            new_node = Node(data)
            p = self.head
            while p.next != None:
                p = p.next
            p.next = new_node
            p = new_node
        return self.head
    
    def middle(self, head):
        slow = head
        fast = head.next
        while fast is not None and fast.next is not None:
            fast = fast.next.next
            slow = slow.next
        return slow

    def reverse(self, head):
        Next = prev = None
        curr = head
        while curr != None:
            Next = curr.next
            curr.next = prev
            prev = curr
            curr = Next
        return prev

    def check_plaindrome(self):
        if self.head is None or self.head.next is None:
            return True
        else:
            mid = self.middle(self.head)
            second = self.reverse(mid.next)
            first = self.head
            is_palindrome = True
            while second:
                if first.data != second.data:
                    is_palindrome = False
                    break
                first = first.next
                second = second.next
            return is_palindrome
        
    def display(self):
        if self.head == None:
            return "empty list."
        else:
            p = self.head
            while p != None:
                print(p.data,end="->")
                p = p.next
            print("None")

ll = LinkedList()
for i in range(int(input("length of linked list"))):
    ll.insert_at_end(int(input("enter linked list element")))
print(ll.check_plaindrome())
ll.display()
```
