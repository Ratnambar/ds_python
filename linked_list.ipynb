```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            return self.head
        else:
            p = self.head
            while p.next != None:
                p = p.next
            p.next = new_node
            return self.head

    def display_list(self):
        if self.head == None:
            return "linked list is empty."
        else:
            temp = self.head
            while temp != None:
                print(temp.data,end="->")
                temp = temp.next
            print("None")

def find_intersection_point_of_two_linked_lists(headA, headB):
    if not headA or not headB:
        return None

    # Function to get the length of a linked list
    def get_length(node):
        length = 0
        while node:
            length += 1
            node = node.next
        return length

    # Get the lengths of both linked lists
    lengthA = get_length(headA)
    lengthB = get_length(headB)

    # Move the longer list's head forward to make them same length
    while lengthA > lengthB:
        headA = headA.next
        lengthA -= 1
    while lengthB > lengthA:
        headB = headB.next
        lengthB -= 1

    # Move both pointers forward until they meet
    while headA.data != headB.data:
        headA = headA.next
        headB = headB.next

    return headA.data

ll1 = LinkedList()
ll2 = LinkedList()
for i in range(int(input("Length of linked list 1: "))):
    ll1.insert_at_end(int(input("Enter linked list 1 data: ")))
for i in range(int(input("Length of linked list 2: "))):
    ll2.insert_at_end(int(input("Enter linked list 2 data: ")))

intersection_node = find_intersection_point_of_two_linked_lists(ll1.head, ll2.head)
if intersection_node:
    print("Intersection node value:", intersection_node)
else:
    print("No intersection node found.")
```
